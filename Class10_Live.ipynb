# Knights and Knaves: 2 People

1.   List item
2.   List item



* Two people
* A_says = A != B
* B_says = A == B


```python
# Code- Simple problem with AB = TF as one known solution
truth_values = [False, True]

def knights_and_knaves():
    for A in truth_values:
        for B in truth_values:
            A_says = A != B
            B_says = A == B

            it_works = (A == A_says) and (B == B_says)

            print(A, B, A_says, B_says, it_works)
```

```python
# Testing
knights_and_knaves()
```

# Knights and Knaves: Four People

* Four people
* A_says = A + B + C + D == 2
* B_says = C == D
* C_says = B == C
* D_says = A != B


```python
# Code- harder problem with ABCD = TTFF as one known solution
truth_values = [False, True]

def knights_and_knaves2():
    for A in truth_values:
        for B in truth_values:
            for C in truth_values:
                for D in truth_values:
                    A_says = A + B + C + D == 2
                    B_says = C == D
                    C_says = B == C
                    D_says = A != B

                    it_works = (A == A_says) and (B == B_says) and (C == C_says) and (D == D_says)

                    print(f'{A}\t{B}\t{C}\t{D}\t{A_says}\t{B_says}\t{C_says}\t{D_says}\t{it_works}')
knights_and_knaves2()
```

```python
# Testing
knights_and_knaves2()
```

# Simultaneous Equations
* 2x - y = 5
* x + 3y = 6


```python
# Code: Similar solution to Knights and Knaves for simultaneous equations

def solve(max_x, max_y):
    for x in range(max_x):
        for y in range(max_y):
            eq1 = (2 * x - y) == 5
            eq2 = (x + 3 * y) == 6
            if eq1 and eq2:
                print(f'({x=}, {y=}) satisfies both equations!')
```

```python
# Testing
solve(10,10)
```

# Removing noise from a sine wave

```python
# Code
import math

# Time series
t = [t/10. for t in range(500)]
print(t)

# The signal
s = [math.sin(val) for val in t]
print(s)
```

```python
# Create a noisy copy!
import random
s_corrupted = [x + 0.05 * random.randint(-10, 10) for x in s]
```

```python
# Try to smooth
def smooth(vector):
    for i in range(1, len(vector)-2):
        vector[i] = (vector[i-1] + vector[i] + vector[i+1]) / 3

vector = s_corrupted[:]

smooth(vector)
smooth(vector)
smooth(vector)
smooth(vector)
smooth(vector)
```

```python
# Plotting three things
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))

plt.subplot(311)
plt.plot(t,s)

plt.subplot(312)
plt.plot(t,s_corrupted)

plt.subplot(313)
plt.plot(t,vector)
plt.show()
```
