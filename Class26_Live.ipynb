```python
# Multiprocessing using terminate
import multiprocessing
import time

def hello():
    for i in range(40):
        print('hello')
        time.sleep(0.25)

def task(name, delay):
    while True:
        print(f'{name}', end='')
        time.sleep(delay)

#task('1', 0.25)

processes = [
    multiprocessing.Process(target=task, args=("1", 0.4)),
    multiprocessing.Process(target=task, args=("2", 0.5)),
    multiprocessing.Process(target=hello, args=()),
    multiprocessing.Process(target=task, args=("3", 1.0)),
]

for p in processes:
    p.start()

input('Hit return when done')

for p in processes:
    p.terminate()



```

```python
# Multiprocessing using join
import multiprocessing
import time

def task(name, delay, number_of_loops):
    for i in range(number_of_loops):
        print(f'{name}', end='')
        time.sleep(delay)

#task('1', 0.25)

processes = [
    multiprocessing.Process(target=task, args=("1", 0.4, 20)),
    multiprocessing.Process(target=task, args=("2", 0.5, 20)),
    multiprocessing.Process(target=task, args=("3", 0.5, 12)),
    multiprocessing.Process(target=task, args=("4", 0.4, 40)),
    multiprocessing.Process(target=task, args=("5", 1.0, 20)),
    multiprocessing.Process(target=task, args=("6", 0.3, 25)),
    multiprocessing.Process(target=task, args=("7", 0.1, 35)),
]

print('Starting...')
for p in processes:
    p.start()
print('Started!')

print('Waiting for all to complete...')
for p in processes:
    p.join()
print('Done!')

```

```python
# Shared Data using multiprocessor.Value
import ctypes

def task(name, delay):
    while True:
        print(f'{name}[{command.value}]', end='')
        if command.value > 0:
            command.value -= 1
        time.sleep(delay)

def linewrapper(name, delay):
    while True:
        print()
        time.sleep(delay)

command = multiprocessing.Value(ctypes.c_int, 0)
processes = [
    multiprocessing.Process(target=task, args=("1", 0.4)),
    multiprocessing.Process(target=task, args=("2", 0.5)),
    multiprocessing.Process(target=task, args=("3", 0.3)),
    multiprocessing.Process(target=task, args=("3", 0.3)),
    multiprocessing.Process(target=linewrapper, args=("wrap", 2)),
]

for p in processes:
    p.start()

while True:
    x = input('Enter new value (q to quit)')
    if x=='q':
        break
    try:
        command.value = int(x)
        print(f'mainline just changed command.value to {command.value}')
    except:
        pass

for p in processes:
    p.terminate()
```
