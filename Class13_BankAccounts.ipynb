```python
# Example: This is generated by ChatGPT
# Show me a Python example using classes and inheritance
# How about something using inheritance and banking
class Account:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}")
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.balance


class Customer:
    def __init__(self, name, accounts=None):
        self.name = name
        self.accounts = accounts if accounts else []
        print(f'Customer {name} created')

    def add_account(self, account):
        self.accounts.append(account)

    def get_total_balance(self):
        a =(account.get_balance() for account in self.accounts)
        #print(type(a))
        total_balance = sum(a)
        return total_balance


# Create accounts
account1 = Account("12345", 1000)
account2 = Account("67890", 500)

# Create a customer and associate accounts
customer1 = Customer("Alice", [account1, account2])

# Deposit and withdraw from accounts
account1.deposit(200)
account2.withdraw(100)

# Print total balance for the customer
print(f"Total balance for {customer1.name}: {customer1.get_total_balance()}")
```

```python
# Now we'll make some child classes to create types of acocunts
class CheckingAccount(Account):
    def __init__(self, account_number, balance=0):
        super().__init__(account_number, balance)
        print(f'Checking Account {account_number} created')

class SavingsAccount(Account):
    def __init__(self, account_number, balance=0):
        super().__init__(account_number, balance)
        print(f'Savings Account {account_number} created')

account1 = SavingsAccount("a12345", 1000)
account2 = CheckingAccount("a67890", 500)
customer1 = Customer("Alice", [account1, account2])

account3 = SavingsAccount("mh12345", 1000)
account4 = CheckingAccount("mh67890", 500)
customer2 = Customer("Mad Hatter", [account3, account4])

customers = [customer1, customer2]

# Deposit and withdraw from accounts
account1.deposit(200)
account2.withdraw(100)
account3.withdraw(130)
account4.withdraw(120)

# Print total balance for the customers
for customer in customers:
    print(f"Total balance for {customer.name}: {customer.get_total_balance()}")
```
